```python
import os
import json
import pandas as pd
```

```python
list_of_json_files = []

directory = 'data/chat-gpt'
 
for filename in os.listdir(directory):
     if filename.endswith('.json'):
#         print(filename)
        list_of_json_files.append(filename)
        
# print(list_of_json_files)

data_list = []

for file in list_of_json_files:
#     print(type(file))
#     with open('data/chat-gpt/20221201.json') as f:
    with open(directory + "/" + file) as f:
        for line in f:
            data_list.append(json.loads(line))

# print(data_list)

df = pd.DataFrame()

# print(type(data_list))

df = pd.DataFrame.from_records(data_list)

df
```
